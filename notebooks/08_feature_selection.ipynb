```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))

import joblib
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
```

```python
DATA_PATH = PROJECT_ROOT / "data" / "processed"

X_train = pd.read_csv(DATA_PATH / "X_train.csv")
X_test = pd.read_csv(DATA_PATH / "X_test.csv")

y_train = pd.read_csv(DATA_PATH / "y_train.csv")["Label"]
y_test = pd.read_csv(DATA_PATH / "y_test.csv")["Label"]

print(X_train.shape)
print(X_test.shape)
```

```python
Selected= [
    "OriginalSNR",
    "AmplitudeStd",
    "AmplitudeKurtosis",
    "PhaseStd",
    "FrequencyStd",
    "SigmaAF",
    "SigmaDP",
    "SigmaZ2",
    "BlindSNR",
    "M2M4SNR",
    "SpectralEntropy",
    "SpectralBandwidth",
    "SpectralSymmetry",
    "M20",
    "C42",
    "R1",
    "R2",
    "R3",
    "R4",
    "R5"
]
```

```python
X_train = X_train[Selected]

X_test = X_test[Selected]

print(X_train.shape)
print(X_test.shape)
```

```python
scaler = StandardScaler()

X_train_scaled = pd.DataFrame(
    scaler.fit_transform(X_train),
    columns=Selected
)

X_test_scaled = pd.DataFrame(
    scaler.transform(X_test),
    columns=Selected
)
```

```python
OUTPUT = PROJECT_ROOT / "data" / "processed"

OUTPUT.mkdir(
    parents=True,
    exist_ok=True
)

pd.Series(
    Selected,
    name="Feature"
).to_csv(
    OUTPUT / "selected_features.csv",
    index=False
)

joblib.dump(
    list(Selected),
    OUTPUT / "selected_features.pkl"
)

joblib.dump(
    scaler,
    OUTPUT / "standard_scaler.pkl"
)

X_train.to_csv(
    OUTPUT / "X_train.csv",
    index=False
)

X_test.to_csv(
    OUTPUT / "X_test.csv",
    index=False
)

X_train_scaled.to_csv(
    OUTPUT / "X_train_scaled.csv",
    index=False
)

X_test_scaled.to_csv(
    OUTPUT / "X_test_scaled.csv",
    index=False
)

print("Files Saved Successfully")
```
