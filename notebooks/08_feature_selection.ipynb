# 08 Feature Selection (CPU Optimized)

```python
from pathlib import Path
import sys
PROJECT_ROOT=Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path: sys.path.append(str(PROJECT_ROOT))
import joblib,numpy as np,pandas as pd,matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold,SelectFromModel
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv(PROJECT_ROOT/'data'/'processed'/'feature_dataset.csv')
X=df.drop(columns=['Modulation','Label'])
y=df['Label']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
vt=VarianceThreshold()
X_train=pd.DataFrame(vt.fit_transform(X_train),columns=X_train.columns[vt.get_support()])
X_test=X_test[X_train.columns]
```

```python
selected=['Std', 'Maximum', 'Minimum', 'Median', 'FFTMean', 'FFTStd', 'FFTMaximum', 'DominantFrequency', 'SpectralCentroid', 'SpectralSpread', 'IMean', 'QMean', 'IStd', 'QStd', 'IQCorrelation', 'PhaseStd']
```

```python
X_train=X_train[selected]
X_test=X_test[selected]
scaler=StandardScaler()
X_train_scaled=pd.DataFrame(scaler.fit_transform(X_train),columns=selected)
X_test_scaled=pd.DataFrame(scaler.transform(X_test),columns=selected)
out=PROJECT_ROOT/'data'/'processed'
out.mkdir(parents=True,exist_ok=True)
pd.Series(selected,name='Feature').to_csv(out/'selected_features.csv',index=False)
joblib.dump(selected,out/'selected_features.pkl')
joblib.dump(scaler,out/'standard_scaler.pkl')
X_train_scaled.to_csv(out/'X_train_scaled.csv',index=False)
X_test_scaled.to_csv(out/'X_test_scaled.csv',index=False)
y_train.to_frame('Label').to_csv(out/'y_train.csv',index=False)
y_test.to_frame('Label').to_csv(out/'y_test.csv',index=False)
print(selected)
print('Done')
```
